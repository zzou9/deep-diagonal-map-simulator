```python
import numpy as np # math operations
import pandas as pd # handling data
import matplotlib.pyplot as plt # graphing
from mpl_toolkits.mplot3d import Axes3D
```

```python
url_coords = '/Users/zou/Documents/brown/codes/pentagram/out/traj_coords/traj_coords_2.csv'
df_coords = pd.read_csv(url_coords)
```

```python
# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df_coords['x0'], df_coords['x1'], df_coords['x2'], c=df_coords['x2'], cmap='viridis', marker='o', s=1)

# Add color bar
colorbar = plt.colorbar(sc, ax=ax)
colorbar.set_label('Color Intensity')

# Add labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot with Color Bar')

# Show the plot
plt.show()
```
